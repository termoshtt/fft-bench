```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./result.csv").set_index("n")
```

```python
df.plot(marker="+")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("array size", fontsize=18)
plt.ylabel("Elapsed Time (ns)", fontsize=18)
plt.legend(fontsize=15)
plt.savefig("result.png")
```

```python
plt.plot(df.index, df["chfft"].values / df["fftw"].values, "x-", label="chfft")
plt.plot(df.index, df["fftw"].values / df["fftw"].values, "+-", label="fftw")
plt.plot(df.index, df["rustfft"].values / df["fftw"].values, "+-", label="rustfft")

plt.xscale("log")
plt.xlabel("array size", fontsize=18)
plt.ylabel("Ratio against FFTW", fontsize=18)
plt.legend(fontsize=15, ncol=3)
plt.savefig("ratio.png")
```
